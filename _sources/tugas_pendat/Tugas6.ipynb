# Decision Tree

Decision tree merupakan suatu struktur yang digunakan untuk membantu proses pengambilan keputusan. Disebut sebagai “tree” karena struktur ini menyerupai sebuah pohon lengkap dengan akar, batang, dan percabangannya. Dalam data science, struktur decision tree dapat membantu ambil keputusan efektif dan tetap memperhatikan kemungkinan hasil serta konsekuensinya

## Konsep Decision Tree

Decision tree merupakan sebuah algoritma yang dapat digunakan untuk klasifikasi dan regresi. Kelebihan menggunaka Decision Tree

1.   Sederhana untuk dipahami dan diimplementasikan
2.   Membutuhkan sedikit persiapan data

Kekurangan menggunakan Decision Tree :



1.   Prediksi Decision tree tidak atau kontinu.
2.   Decision Tree dapat menjadi tidak stabil karena variasi kecil dalam data.






```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn import tree
from matplotlib import pyplot as plt
```

```python
# Reading the Iris.csv file
df = pd.read_csv("https://raw.githubusercontent.com/rizkyluxszerr/datamining/main/iris_csv.csv")
df.head()
```

```python
y = df["class"]
X = df.drop(columns=["class"])
clf = tree.DecisionTreeClassifier(criterion="gini")
clf = clf.fit(X, y)
```

```python
fig = plt.figure(figsize=(10,10))
_ = tree.plot_tree(clf, feature_names=list(df.columns.values)[:4], class_names=list(df.columns.values)[4] ,filled=True)
```
