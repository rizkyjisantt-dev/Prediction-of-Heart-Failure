### **Mean / Rata-Rata**
Dalam statistika dan matematika, mean disebut juga dengan rata-rata. Jadi, definisi mean adalah nilai rata-rata dari sekumpulan data yang terdiri dari dua atau lebih..
Berikut rumus mean untuk data tunggal:
$$\bar{x}=\frac{\sum x_i}{n}
$$
Keterangan:

$\bar{x}:$ nilai rata-rata

$\mathrm{x}_{\mathrm{i}}$ : nilai data ke-i

n : banyaknya data

````python
#menampilkan rata-rata
st.mean(df.jumlah_korban)
````

### **Modus**

Modus adalah nilai data yang paling sering muncul atau nilai data yang frekuensinya paling besar.

````python
#menampilkan rata-rata
st.mode(df.jumlah_korban)
````

### **Median**

Median disebut juga dengan nilai tengah. Median diperoleh dengan cara mengambil nilai tengah pada data yang telah disusun secara berurutan dari angka terkecil ke terbesar.
Pada data tunggal untuk menentukan median yaitu:
Misalkan banyaknya data n, 

Untuk $n$ ganjil: $M e=x_{\frac{1}{2}(n+1)}$

Untuk $n$ genap: $M e=\frac{\frac{x_n}{2}+\frac{x_n}{2}+1}{2}$

Keterangan: 
$x_{\frac{n}{2}}=$ data pada urutan ke- $\frac{n}{2}$ setelah diurutkan.

````python
#menampilkan Median dari data
st.median(df.jumlah_korban)
````

### **Kuartil**
Kuartil adalah suatu rumus yang membagi suatu data menjadi kepada empat yang sama banyak. Kemudian dari setiap data yang terbagi sama banyak tersebut dibatasi oleh sebuah nilai.

Seperti Pada kuartil, misalakan empat data yang dibagi menjadi sama banyak, akan dibatasi oleh 3 tiga nilai kuartil yaitu kuartil atas, kuartil tengah, dan kuartil bawah.

````python
#menampilkan kuartil dari data
np.percentile(df.jumlah_korban, [25,50,75])
````

### **Range(Jangkauan)**

Dalam sekelompok data kuantitatif akan terdapat data dengan nilai terbesar dan data dengan nilai terkecil. Rentang (range) atau disebut juga dengan jangkauan adalah selisih antara data dengan nilai yang terbesar dengan data denga nilai yang terkecil tersebut.

$$R = x_{max} - x_{min}$$

dimana R adalah range (jangkauan atau rentang), x_{max}x 
max
​
  adalah nilai data yang paling besar dan x_{min}x 
min
​
  nilai data yang paling kecil.

````python
# menampilkan Jangkaun dari data
# rumus jangkauan J = Xn - X1
Xn = max(df['jumlah_korban'])
X1 = min(df['jumlah_korban'])
J = Xn - X1
J
````
